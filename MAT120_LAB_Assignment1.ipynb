```python
#1

from sympy import*

x = symbols('x')

y = 9*x**3 - 7*x**2 + 3*x + 10

df = diff(y, x)

f = lambdify(x, y)

from numpy import*
p = arange(-5, 6)
r = f(p)

import matplotlib.pyplot as plt
plt.plot(p, r)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('9*x**3 - 7*x**2 + 3*x + 10')
plt.show()

a = solve(df, x)

minima = y.subs(x, a[0])
maxima = y.subs(x, a[1])
print('Minima:  ', minima)
print('Maxima:  ', maxima)

#complete
```
